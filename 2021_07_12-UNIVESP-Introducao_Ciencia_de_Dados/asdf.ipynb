```python
import pandas as pd
import numpy as np
import time
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

length = 100000
df = pd.DataFrame()
df['Pessoas'] = np.random.randint(190, 2059, size=length)
df['Compra'] = np.random.choice(['A', 'B', 'C', 'D'], length)

def filtro1():
  b = df.query('Compra=="B" & Pessoas=="2020" ')
  return b 

def filtro6():
  b = df[(df['Compra']=='B') & (df['Pessoas']=='2020')]
  return b

def filtro7(df):
  b = df[(df['Compra']=='B') & (df['Pessoas']=='2020')]
  return b

tempo1 = []
tempo2 = []
tempo3 = []
tempo4 = []
tempo5 = []
tempo6 = []
tempo7 = []
tempo8 = []
tempo9 = []
lenghts = []

for i in np.arange(0, 100):
  length = 100000 * i
  df = pd.DataFrame()
  df['Pessoas'] = np.random.randint(190, 2059, size=length)
  df['Compra'] = np.random.choice(['A', 'B', 'C', 'D'], length)
  s = time.clock()
  result = df.query('Compra=="B" and Pessoas=="2020" ')
  e = time.clock()
  tempo1.append(np.round((e - s),5))
  s = time.clock()
  result = df.query('Compra=="B" & Pessoas=="2020" ')
  e = time.clock()
  tempo2.append(np.round((e - s),5))
  s = time.clock()
  result = df.query('Compra=="B" & Pessoas==2020 ')
  e = time.clock()
  tempo3.append(np.round((e - s),5))
  s = time.clock()
  result = df.query('Compra=="B" and Pessoas==2020 ')
  e = time.clock()
  tempo4.append(np.round((e - s),5))
  s = time.clock()
  result = df[(df['Compra']=='B') & (df['Pessoas']==2020)]
  e = time.clock()
  tempo5.append(np.round((e - s),5))
  s = time.clock()
  result = df[(df['Compra']=='B') & (df['Pessoas']=='2020')]
  e = time.clock()
  tempo6.append(np.round((e - s),5))

for i in np.arange(0, 100):
  length = 100000 * i
  lenghts.append(length)
  df = pd.DataFrame()
  df['Pessoas'] = np.random.randint(190, 2059, size=length)
  df['Compra'] = np.random.choice(['A', 'B', 'C', 'D'], length)
  s = time.clock()
  result = filtro6()
  e = time.clock()
  tempo7.append(np.round((e - s),5))
  s = time.clock()
  result = filtro7(df)
  e = time.clock()
  tempo8.append(np.round((e - s),5))
  s = time.clock()
  result = filtro1()
  e = time.clock()
  tempo9.append(np.round((e - s),5))

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=lenghts,y=tempo1,
                    mode='lines',
                    name='Primeira var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo2,
                    mode='lines',
                    name='Segunda var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo3,
                    mode='lines',
                    name='Terceira var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo4,
                    mode='lines',
                    name='Quarta var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo5,
                    mode='lines',
                    name='Quinta var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo6,
                    mode='lines',
                    name='Sexta var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo7,
                    mode='lines',
                    name='Sétima var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo8,
                    mode='lines',
                    name='Oitava var'))
fig.add_trace(go.Scatter(x=lenghts,y=tempo9,
                    mode='lines',
                    name='Nona var'))
fig.update_layout(title='Tempo de processamento por técnicas diferentes de filtro',
                   xaxis_title='Tamanho da base de dados',
                   yaxis_title='Tempo de processamento')

fig.show()
```
